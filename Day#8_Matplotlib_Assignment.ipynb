```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
# Create charts using list
fruit = ['Apple','Banana','Orange', 'Strawberry',]
sales = [20 , 30, 10 ,40 ]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sales,labels=fruit, autopct='%1.1f%%', colors =colors,  startangle=180, radius=1.5)
plt.show()
```

```python
#Create charts using dataframe
data = {'year': [2010, 2011, 2012, 2013],
        'team': ['FCBarcelona', 'RMadrid', 'ValenciaCF', 'temeA'],
        'wins':   [30, 28, 32, 29],
        'draws':  [6, 7, 4, 5],
        'losses': [2, 3, 2, 4]}

football = pd.DataFrame(data, columns=['year', 'team', 'wins', 'draws', 'losses'])  

plt.xlabel('Team')
plt.ylabel('wins')
plt.title('Number of wins for each team ')

plt.plot(football['team'],football['wins'], c = '0.4')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
#Number of losses for each team
plt.pie(football['losses'],labels=football['team'], autopct='%1.1f%%', colors =colors,  startangle=180, radius=1.5)
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(football['team'],football['draws']) 

plt.title('draes number',fontsize = 20)
plt.xlabel('team',fontsize = 20)
plt.ylabel('draws',fontsize = 20)

plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(football['team'],football['draws']) 

plt.title('draes number',fontsize = 20,ha = 'right', color='red' )
plt.xlabel('team',fontsize =16 )
plt.ylabel('draws',fontsize = 16)

plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot(football['team'],football['losses'], c ='green', linestyle = 'solid', linewidth = 3)
plt.plot(football['team'],football['wins'], c ='red', linestyle = 'solid', linewidth = 3)
plt.plot(football['team'],football['draws'], c ='black', linestyle = 'solid', linewidth = 3)


plt.legend(['wins','losses','draws'], shadow = True, loc = 0) # location (0 = best)
plt.xlabel('team')
plt.ylabel('year')
plt.title('Number of wins,losses and draws of each team')
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

#first
plt.subplot(1,3,1) #(number of rows, number of columns, number of plot)
plt.pie(football['losses'],labels=football['team'], autopct='%1.1f%%')
plt.title('losses');

#second
plt.subplot(1,3,2)
plt.scatter(football['team'],football['draws']) 
plt.title('draes number',fontsize = 20 )
plt.xlabel('team',fontsize =16 )
plt.ylabel('draws',fontsize = 16)

#thered
plt.subplot(1,3,3)
plt.plot(football['team'],football['wins'], c = '0.4')
plt.title('Number of wins for each team ')
```
